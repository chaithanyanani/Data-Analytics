## increase and decrease cases

```python
df2 = pd.read_csv("Indian_Crime_Data_Pivoted_2020_2025.csv")

# Define property crimes
property_crimes = ['Arson', 'Burglary', 'Theft', 'Trespassing']

# Replace NaN with 0 and compute total property crimes
df2[property_crimes] = df2[property_crimes].fillna(0)
df2['Total_Property_Crimes'] = df2[property_crimes].sum(axis=1)

# Group by year and state, sum the crimes
grouped = df2.groupby(['Year', 'State'])['Total_Property_Crimes'].sum().reset_index()

# Get unique years in sorted order
years = sorted(df2['Year'].unique())

# Plot one graph per year
for year in years:
    data = grouped[grouped['Year'] == year].sort_values('Total_Property_Crimes', ascending=False)

    states = np.array(data['State'])
    crimes = np.array(data['Total_Property_Crimes'])

    plt.figure(figsize=(12, 8))
    plt.barh(states, crimes, color='mediumvioletred')
    plt.xlabel('Total Property Crimes')
    plt.title(f'Property Crimes by State in {year}')
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()
```

### 5. Which States Had Highest Crime Increase Over Time

```python
df1_pivot = df1.pivot(index='State', columns='Year', values='Total_Crime')
df1_pivot['Increase_%'] = ((df1_pivot[df1['Year'].max()] - df1_pivot[df1['Year'].min()]) / df1_pivot[df1['Year'].min()]) * 100
top_increased_states = df1_pivot['Increase_%'].sort_values(ascending=False).dropna().head(5)

print("ðŸ”¥ Top 5 states with highest crime increase percentage in df1:")
print(top_increased_states)
```

```python
# Pivot df2: States as rows, Years as columns
df2_pivot = df2.pivot(index='State', columns='Year', values='Total_Crime')
# Get first and last year in the dataset
start_year = df2['Year'].min()
end_year = df2['Year'].max()
# Calculate % increase in crime from start to end year
df2_pivot['Increase_%'] = ((df2_pivot[end_year] - df2_pivot[start_year]) / df2_pivot[start_year]) * 100
# Sort and get top 5 states
top_increased_states = df2_pivot['Increase_%'].sort_values(ascending=False).dropna().head(5)
# Display results
print("ðŸ”¥ Top 5 states with highest crime increase percentage in df2:")
print(top_increased_states)
```

```python
df1['Total_Crime'] = df1.drop(columns=['State', 'Year']).fillna(0).sum(axis=1)
df2['Total_Crime'] = df2.drop(columns=['State', 'Year']).fillna(0).sum(axis=1)

# Get latest year data from each dataset
df1_latest = df1[df1['Year'] == df1['Year'].max()]
df2_latest = df2[df2['Year'] == df2['Year'].max()]

# Merge datasets on state
merged = pd.merge(df1_latest, df2_latest, on='State', suffixes=('_df1', '_df2'))

# Calculate change in total crime
merged['Change'] = merged['Total_Crime_df2'] - merged['Total_Crime_df1']

# Filter states with a decrease in crime
crime_decrease = merged[merged['Change'] < 0][['State', 'Change']].sort_values('Change')

# Plotting
plt.figure(figsize=(12, 8))
plt.barh(crime_decrease['State'], crime_decrease['Change'], color='seagreen')
plt.xlabel('Decrease in Total Crimes')
plt.title('States with Decrease in Total Crime (2020 vs 2025)')
plt.tight_layout()
plt.show()
```

## 8.Percentage of Each Crime Type

```python
df1 = pd.read_csv("Indian_Crime_Data_Pivoted_2015_2020.csv")

# Identify only individual crime columns (exclude 'State', 'Year', 'Total_Crime' if it exists)
crime_columns = [col for col in df1.columns if col not in ['State', 'Year', 'Total_Crime']]

# Calculate total for each crime type
crime_totals = df1[crime_columns].sum()

# Calculate percentage
crime_percent = (crime_totals / crime_totals.sum()) * 100

# Plot pie chart
plt.figure(figsize=(10, 10))
plt.pie(crime_percent, labels=crime_percent.index, autopct='%1.1f%%', startangle=140, shadow=True)
plt.title("Crime Type Distribution in df1 (2015â€“2020)", fontsize=14)
plt.axis('equal')  # Ensures pie is a circle
plt.tight_layout()
plt.show()
```

### 10.Statewise % Change in Crime (2020â€“2025 vs 2015â€“2020)

```python
crime_cols = list(set(df1.columns) & set(df2.columns) - {'State', 'Year'})

# Group by state and sum crime data
statewise_2015 = df1.groupby("State")[crime_cols].sum()
statewise_2020 = df2.groupby("State")[crime_cols].sum()
statewise_2015["Total_Crimes_2015_2020"] = statewise_2015.sum(axis=1)
statewise_2020["Total_Crimes_2020_2025"] = statewise_2020.sum(axis=1)
# Merge on state
statewise_compare = pd.merge(
    statewise_2015[["Total_Crimes_2015_2020"]],
    statewise_2020[["Total_Crimes_2020_2025"]],
    left_index=True,
    right_index=True
)
```

```python
# Calculate changes
statewise_compare["Change_in_Crime"] = statewise_compare["Total_Crimes_2020_2025"] - statewise_compare["Total_Crimes_2015_2020"]
statewise_compare["%_Change"] = (statewise_compare["Change_in_Crime"] / statewise_compare["Total_Crimes_2015_2020"].replace(0, pd.NA)) * 100
statewise_compare["Change_in_Crime"]

statewise_compare.sort_values("%_Change", ascending=False)["%_Change"].plot(kind='bar', figsize=(14, 6), color='purple')
plt.title("Statewise % Change in Crime (2020â€“2025 vs 2015â€“2020)")
plt.ylabel("Percentage Change")
plt.xlabel("State")
plt.xticks(rotation=90)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
